r: leitura

w: escrita

a: anexar ao final do arquivo (append)

r+: leitura e escrita

### Lendo arquivos

forma errada:

```python
arquivo = open("/home/felipe/Desktop/data_engineering_journey/data/ex.txt", "r", encoding='utf-8')
conteudo = arquivo.read()
print(conteudo)
```

o problema desse metodo acima é se esquecer de fechar o arquivo, ocupa espaço na memória

```python
with open("/home/felipe/Desktop/data_engineering_journey/data/ex.txt", "r", encoding='utf-8') as arquivo:
    conteudo = arquivo.read()
    print(conteudo)
```

ja o with garante que mesmo o arquivo com erro, vai fechar ao final

### Escrevendo arquivos

```python
with open("/home/felipe/Desktop/data_engineering_journey/data/ex2.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write("Teste 123\n")
    arquivo.write("Teste 345\n")
```

### Adicionando ao final do arquivo (append)

```python
with open("/home/felipe/Desktop/data_engineering_journey/data/ex.txt", "a", encoding="utf-8")as arquivo:
    arquivo.write("\n678")
```

### Percorrendo pelo arquivo no terminal

```python
with open("/home/felipe/Desktop/data_engineering_journey/data/ex3.txt", "r", encoding="utf-8")as arquivo:
    for linha in arquivo:
        print(linha.strip())
```
