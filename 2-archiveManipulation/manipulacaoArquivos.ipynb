### Leitura de linhas especificas com enumerate

```python
with open("/home/felipe/Desktop/data_engineering_journey/data/ex3.txt", "r+", encoding="utf-8") as arquivo:
    for index, linha in enumerate(arquivo):
        if index < 2:
            print(linha.strip())   #strip é só pra remover espaço extra nas linhas
```

### Posicionando o cursor na operaçao de escrita também

```python
with open("/home/felipe/Desktop/data_engineering_journey/data/ex.txt", "r+", encoding="utf-8") as arquivo:
    arquivo.seek(10)
    arquivo.write("texto novo")
```

### Criando um loop infinito que percorre um bloco e so para quando nao tem mais nada

```python
with open("/home/felipe/Desktop/data_engineering_journey/data/ex3.txt", "r+", encoding="utf-8") as arquivo:
    while True:
        print("---INICIO DO BLOCO---")
        bloco = arquivo.read(10)
        
        if not bloco:
            break
        
        
        print(bloco)
        print("---FINAL DO BLOCO---")
```
