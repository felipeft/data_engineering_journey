Filtro para livros em estoque apenas

```python
import json

with open("/home/felipe/Desktop/data_engineering_journey/data/livros.json", "r", encoding="utf=8") as arquivo:
    livros = json.load(arquivo)

print("Livros disponiveis para compra")
for livro in livros:
    if livro["em_estoque"]:
        print(f"{livro["titulo"]} - R${livro["preco"]:.2f}")
```

Adicionando novo livro

```python
novo_livro = {
    "id": 3,
    "titulo": "Duna",
    "autor": "Frank Herbert",
    "preco": 65.00,
    "em_estoque": True
}

livros.append(novo_livro)
print(livros)
```

### criando um novo json de livros atualizados

```python
with open("/home/felipe/Desktop/data_engineering_journey/data/livros_atualizados.json", "w", encoding="utf-8")as arquivo:
    json.dump(livros, arquivo, indent=4, ensure_ascii=False)

#indent e ensure_ascii pra deixar formatado no padrao (4 de espaço) e com acentos (ensure_ascii)
```

### exemplo como se tivesse enviando ou consumindo json em api
vem em formato de texto geralmente

```python
dados_aluno = {"nome":"João", "nota": 9.5, "aprovado": True}

texto_json = json.dumps(dados_aluno, ensure_ascii=False)  #dump(s) para strings que estamos passando

print(type(texto_json))


## geralmente é asism que envia pra api
```

```python
resposta_da_api = '{"curso": "Python", "modulos": 10, "ativo": true}'
dicionario = json.loads(resposta_da_api)
print(dicionario)
print(dicionario["curso"])
print(type(dicionario))
```
