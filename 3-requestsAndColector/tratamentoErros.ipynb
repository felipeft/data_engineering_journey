### "Aprendendo com os erros"

para tratar erros em python, utiliza-se a clausula try, caso tenha um erro no try, resolvemos no except

fazendo um "Somador automatico"

```python
n1 = int(input("Digite um número: "))
n2 = int(input("Digite outro número: "))
resultado = n1 + n2

print(f"A soma dos dois é: {resultado}")
```

agora para gerar um erro vou por um caractere em vez de numero
isso vai dar o seguinte erro 
ValueError: invalid literal for int() with base 10: ''

fazendo agora com tratamento de erro

```python
try:
    n1 = int(input("Digite um número: "))
    n2 = int(input("Digite outro número: "))
    resultado = n1 + n2

    print(f"A soma dos dois é: {resultado}")
except:
    print("Digite um número válido")
```

também podemos reescrever dessa foma, onde agora o proprio erro será colocado como um objeto e poderemos até printar na tela

```python
try:
    n1 = int(input("Digite um número: "))
    n2 = int(input("Digite outro número: "))
    resultado = n1 + n2

    print(f"A soma dos dois é: {resultado}")
except Exception as erro:
    print(erro)
    print("Digite um número válido")
```

Ou seja, Todos os erros vem da classe Exception
Vamos ver outros erros:

```python

n1 = float(input("Digite um número: "))
resultado = 10 / n1

print(f"O resultado é: {resultado}")
```

note que agora foi um "ValueError:" pois coloquei um caractere, mas se tentassemos fazer divisao por zero

```python

n1 = float(input("Digite um número: "))
resultado = 10 / n1

print(f"O resultado é: {resultado}")
```

O erro agora "ZeroDivisionError" é diferente. logo para cada erro temos uma classe propria pra ele

então podemos tratar da seguinte forma:

```python
try:
    n1 = float(input("Digite um número: "))
    resultado = 10 / n1

    print(f"O resultado é: {resultado}")
except ValueError as erro:
    print(f"Erro ocorrido: {erro}")
```

beleza, mas se colocassemos um zero:

```python
try:
    n1 = float(input("Digite um número: "))
    resultado = 10 / n1

    print(f"O resultado é: {resultado}")
except ValueError as erro:
    print(f"Erro ocorrido: {erro}")
```

Isso acontece pois o unico erro que ele está capturando é o ValueError, e nao o ZeroDivision e demais.
Ná pratica usando somente a classe Exception ja daria pra reconhecer todos, mas melhor pratica e fazer tratamento para todos os erros

```python
try:
    n1 = float(input("Digite um número: "))
    resultado = 10 / n1

    print(f"O resultado é: {resultado}")
except ValueError as erro:
    print(f"Erro ocorrido: {erro}")
except ZeroDivisionError as erro:
    print(f"Erro ocorrido: {erro}")
```

### Usando o If no try

```python
try:
    n1 = float(input("Digite um número: "))
    resultado = n1 / 10

except ValueError as erro:
    print(f"Erro ocorrido: {erro}")
except ZeroDivisionError as erro:
    print(f"Erro ocorrido: {erro}")
else:                                       # o else é executado quando não da nenhum erro
    print(f"O resultado é: {resultado}")
```

e pra finalizar de vez o programa, independente se deu certo ou não

```python
try:
    n1 = float(input("Digite um número: "))
    resultado = n1 / 10

except ValueError as erro:
    print(f"Erro ocorrido: {erro}")
except ZeroDivisionError as erro:
    print(f"Erro ocorrido: {erro}")
else:                                       # o else é executado quando não da nenhum erro
    print(f"O resultado é: {resultado}")
finally:
    print("programa finalizado")
```

### criando os proprios erros

agora vamos criar os proprios erros no python em vez de solucionar.  
usando o Raise  
apenas para casos especificos  

```python
try:
    idade = int(input("Digite sua idade: "))
except Exception as erro:
    print(f"erro: {erro}")
else:
    print(f"Sua idade é: {idade}")
finally:
    print("Fim do programa")
```

Digitei uma idade negativa e mesmo assim passou de boa  
no python isso tem que ser aceitavel pois é um simples inteiro   
mas numa logica de negocios de um projeto grande pode ter certeza que isso é um bucho  
entao vamos criar um erro para travar isso  

```python
try:
    idade = int(input("Digite sua idade: "))
    if idade < 0:
        raise Exception("Digite uma idade válida!")     # raise é tipo "lançar o erro"
except Exception as erro:
    print(f"erro: {erro}")
else:
    print(f"Sua idade é: {idade}")
finally:
    print("Fim do programa")
```

Mas nesses casos é melhor digitar o tipo de erro que estamos provocando (por isso é bom ler a documentação)  
como por exemplo:

```python
try:
    idade = int(input("Digite sua idade: "))
    if idade < 0:
        raise ValueError("Digite uma idade válida!")     # raise é tipo "lançar o erro"
except Exception as erro:
    print(f"erro: {erro}")
else:
    print(f"Sua idade é: {idade}")
finally:
    print("Fim do programa")
```

Isso funcionou justamente por conta da classe "Exception" no except pois todo erro vai herdar essa classe
